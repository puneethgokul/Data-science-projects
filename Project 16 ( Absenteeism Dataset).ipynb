Outlier treatment

```python
def impute_outliers(col):
    col_median = col.median()
    col_std  = col.std()
    
    return np.where((col - col_median).abs() > 2*col_std ,col_median,col)
```

```python
#Only treating columns that have significantly more outliers
df['Transportation expense'] = impute_outliers(df['Transportation expense'])

df['Work load Average/day '] = impute_outliers(df['Work load Average/day '])
```

preparing the data

```python
X = df.drop('Absenteeism time in hours', axis=1)
# the dependent variable
Y = df['Absenteeism time in hours']
```

svm regressor

```python
from sklearn import svm
 
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

newdict = dict()


clr = svm.SVR()
kfold = KFold(n_splits = 10)
results = cross_val_score(clr, X, Y, cv=kfold)
        
print("model_score = {}%".format(results.mean()*100.0))
```

Desicion tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

newdict = dict()
max_features_list = list(range(1,17))
max_depth_list = list(range(3,10))

for f in max_features_list:
     for d in  max_depth_list:
            model_dt = DecisionTreeRegressor(max_features=f , max_depth = d)
            kfold = KFold(n_splits = 10)
            results = cross_val_score(model_dt, X, Y, cv=kfold)
        
            newdict["max_features= {}, max_depth = {}".format(f,d)] = "model_score = {}%".format(results.mean()*100.0)
            

print(max(newdict,key = newdict.get),max(newdict.values()))
```

Random forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

newdict2 = dict()
max_depth_list = list(range(3,9))
n_estimators_list = list(range(50,500,50))

for d in max_depth_list:
     for n in  n_estimators_list:
            model_rf = RandomForestRegressor(max_depth=d , n_estimators = n)
            kfold = KFold(n_splits = 10)
            results = cross_val_score(model_rf, X, Y, cv=kfold)
        
            newdict2["n_estimators= {}, max_depth = {}".format(n,d)] = "model_score = {}%".format(results.mean()*100.0)
            

print(max(newdict2,key = newdict2.get),max(newdict2.values()))
```
