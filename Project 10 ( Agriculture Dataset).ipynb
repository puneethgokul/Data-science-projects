```python
#target variable

sns.countplot(df_train['Crop_Damage'])
```

```python
x_train = df_train.drop(['ID', 'Crop_Damage'],axis=1)     
y_train = df_train['Crop_Damage']
```

logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

model_lr = LogisticRegression(max_iter=10000)

kfold = KFold(n_splits = 4)

results = cross_val_score(model_lr, x_train, y_train, cv=kfold)

print(results)
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0, results.std()*100.0))
```

K-Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier

scores =[]

for k in range(1,46):
    model_knn = KNeighborsClassifier(n_neighbors = k, weights = 'distance' )
    kfold = KFold(n_splits = 4)
    results = cross_val_score(model_knn, x_train, y_train, cv=kfold)
    
    scores.append(results.mean()*100.0)

plt.plot(range(1,46),scores)
```

Decision tree

```python
from sklearn.tree import DecisionTreeClassifier


newdict = dict()
max_features_list = list(range(1,6))
max_depth_list = list(range(1,7))

for f in max_features_list:
     for d in  max_depth_list:
            model_dt = DecisionTreeClassifier(max_features= f, max_depth = d)
            kfold = KFold(n_splits = 10)
            results = cross_val_score(model_dt, x_train, y_train, cv=kfold)
        
            newdict["max_features= {}, max_depth = {}".format(f,d)] = results.mean()*100.0
            

print(max(newdict,key = newdict.get),max(newdict.values()))
```

Random forest

```python
from sklearn.ensemble import RandomForestClassifier

newdict2 = dict()
max_depth_list = list(range(2,7))
n_estimators_list = [50,100,150,250]

for d in max_depth_list:
     for n in n_estimators_list:
            model_rfcl = RandomForestClassifier(max_depth=d, n_estimators= n)
            kfold = KFold(n_splits = 10)
            results = cross_val_score(model_rfcl, x_train, y_train, cv=kfold)
        
            newdict2["max_depth= {}, n_estimators = {}".format(d,n)] = results.mean()*100.0 
            

print(max(newdict2,key = newdict2.get),max(newdict2.values()))
```

All the models are performing equally good with an 83% accuracy
