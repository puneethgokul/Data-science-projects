```python
#Is the distribution of bmi across women with no children, one child and two children, the same

#classifying 0,1,2 children rows
ins_c0=ins.loc[(ins['children']==0) & (ins['sex']=="female")]
ins_c1=ins.loc[(ins['children']==1) & (ins['sex']=="female")]
ins_c2=ins.loc[(ins['children']==2) & (ins['sex']=="female")]


#plotting all 3 graphs together
sns.distplot(ins_c0['bmi'],color='red') , sns.distplot(ins_c1['bmi'],color='blue'),sns.distplot(ins_c2['bmi'],color='yellow')

plt.ylim(0,0.08)
plt.title('Distribution of BMI of females with o,1 and children',color='green')
```

>above is the distribution of bmi across women with no children(red), one child(blue) and two children(yellow) overlapped.
>All three graphs are normally distributed.
>moreover, as we can see all three distributions almost are similar and is only little part that is not overlapping
